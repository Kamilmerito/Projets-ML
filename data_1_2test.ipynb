**DATA VISUALISATION DATA2**

```python
#création d'une fonction de construction du diagramme et de manière interactive
def bar_plot(a):
  df[a].value_counts(normalize=True).plot(kind="bar")
  plt.ylabel("Proportion")
  plt.title("Distribution of "+str(a))
  return plt.show()
```

```python
#interact
interact(bar_plot, a=categorical_col)
```
