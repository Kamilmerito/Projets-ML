**AUTOMATISATION DE VISUALISATION**

```python
#crÃ©ation d'une fonction de construction d'histogramme  et de maniÃ¨re interactive
def hist_plot(b):
  sns.displot(df[b],kde=False)
  plt.title("Histogram of"+str(b))
  return plt.show()
```

```python
#interact

interact(hist_plot, b=numerical_col);
```

```python
def bar_plot(a):
  df[a].value_counts(normalize=True).plot(kind="bar")
  plt.ylabel("Proportion")
  plt.title("Distribution of "+str(a))
  return plt.show()
```

```python
#interact
interact(bar_plot, a=categorical_col)
```
